```python
column = 'AMT_ANNUITY'
sub_column = app_train[column]

print("column: %s" % column)
print("mean = %f" % sub_column.mean())
print("std = %f" % sub_column.std())
# sub_column.describe()
```

```python
# to fill Nan with value 0
sub_column.fillna(0)
hist_row = np.array(sub_column)

plt.hist(hist_row) 
plt.title("histogram") 
plt.show()
```
