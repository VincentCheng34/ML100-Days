# 檢視與處理 Outliers
### 為何會有 outliers, 常見的 outlier 原因
* 未知值，隨意填補 (約定俗成的代入)，如年齡常見 0,999
* 可能的錯誤紀錄/手誤/系統性錯誤，如某本書在某筆訂單的銷售量 = 1000 本

```python
# Import 需要的套件
import os
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

%matplotlib inline

# 設定 data_path
dir_data = './data'
```

```python
f_app = os.path.join(dir_data, 'application_train.csv')
print('Path of read in data: %s' % (f_app))
app_train = pd.read_csv(f_app)
app_train.head()
```

## 請參考 HomeCredit_columns_description.csv 的欄位說明，觀察並列出三個你覺得可能有 outlier 的欄位並解釋可能的原因

```python
typeList = list(app_train.dtypes)
print(list(set(typeList)))
print(app_train.dtypes.value_counts())
```

```python
# 先篩選數值型的欄位
"""
YOUR CODE HERE, fill correct data types (for example str, float, int, ...)
"""
df_num = app_train.select_dtypes(include=[np.number])

# 再把只有 2 值 (通常是 0,1) 的欄位去掉
for col in df_num:
    if len(np.unique(df_num[col])) == 2:
        del df_num[col]

print(df_num.shape[1])

# # 檢視這些欄位的數值範圍
n_rows = int(np.ceil(df_num.shape[1] / 2))
fig, axes = plt.subplots(n_rows, 2, figsize=(15, n_rows*2))
for (col, ax) in zip(df_num.columns.tolist(), axes.ravel()):
    df_num.boxplot(column=col, ax=ax, vert=False)
    ax.set_yticklabels([])
    ax.set_title(col)
plt.tight_layout()
```

### Hints: Emprical Cumulative Density Plot, [ECDF](https://zh.wikipedia.org/wiki/%E7%BB%8F%E9%AA%8C%E5%88%86%E5%B8%83%E5%87%BD%E6%95%B0), [ECDF with Python](https://stackoverflow.com/questions/14006520/ecdf-in-python-without-step-function)

```python
# 經驗分布函數（英語：empirical distribution function）
def ecdf(data):
    """Compute ECDF for a one-dimensional array of measurements."""
    # Number of data points: n
    n = len(data)

    # x-data for the ECDF: x
    x = np.sort(data)

    # y-data for the ECDF: y
    y = np.arange(1, n+1) / n

    return x, y
```

```python
# 最大值離平均與中位數很遠
print(app_train['AMT_INCOME_TOTAL'].describe())

# 繪製 Empirical Cumulative Density Plot (ECDF)
"""
YOUR CODE HERE
"""
x, y = ecdf(app_train['AMT_INCOME_TOTAL'])
plt.plot(x, y)
plt.xlabel('Value')
plt.ylabel('ECDF')
plt.show()

# 改變 y 軸的 Scale, 讓我們可以正常檢視 ECDF
x, y = ecdf(app_train['AMT_INCOME_TOTAL'])
plt.plot(x, y)
plt.xscale('log')
plt.xlabel('Value (log-scale)')
plt.ylabel('ECDF')
plt.show()
```
