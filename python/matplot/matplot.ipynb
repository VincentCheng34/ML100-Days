```python
# %matplotlib notebook or %matplotlib inline
# %matplotlib inline => 不需每次執行 plt.show()
%matplotlib inline

import matplotlib.pyplot as plt
import numpy as np
```

# 圖與子圖

```python
fig = plt.figure()

ax1 = fig.add_subplot(2,2,1)
ax2 = fig.add_subplot(2,2,2)
ax3 = fig.add_subplot(2,2,3)

ax1.hist(np.random.randn(100), bins=20, color='k', alpha=0.3)
ax2.scatter(np.arange(30), np.arange(30)+3*np.random.randn(30))
ax3.plot(np.random.randn(50).cumsum(), 'k--')
```

# 陣列索引子圖

```python
# axes = plt.subplots(2, 3)
fig, axes = plt.subplots(2, 3)

axes[0, 1].plot(np.random.randn(50).cumsum(), 'k--')
```

# 調整子圖間留白

```python
# subplots_adjust(wspace=0, hspace=0)
fig, axes = plt.subplots(2, 2, sharex=True, sharey=True)
for i in range(2):
    for j in range(2):
        axes[i, j].hist(np.random.randn(500), bins=50, color='k', alpha=0.5)
plt.subplots_adjust(wspace=0, hspace=0)
```

# 色彩、標示和線條樣式




```python
fig, axes = plt.subplots(2, 2)

axes[0, 0].plot(np.random.randn(50).cumsum(), 'ro-')
axes[0, 1].plot(np.random.randn(50).cumsum(), 'g^--')
axes[1, 0].plot(np.random.randn(50).cumsum(), 'b>-.')
axes[1, 1].plot(np.random.randn(50).cumsum(), 'yx:')
```

# 一個圖形多條線、圖例

```python
fig, axes = plt.subplots(1, 1)

data = np.random.randn(30).cumsum()
axes.plot(data, 'k--', label='Default')
axes.plot(data, 'g-', drawstyle='steps-post', label='steps-post')

# 圖例
# 不要圖例 label='_nolegend_'
plt.legend(loc='best')
```

# 刻度、標籤和圖例
plt.xlim(), plt.xticks(), plt.xticklabels()

```python
fig, axes = plt.subplots(1, 2)

data = np.random.randn(30).cumsum()
axes[0].plot(data, 'k--', label='Default')
axes[0].plot(data, 'g-', drawstyle='steps-post', label='steps-post')

axes[1].plot(data, 'k--', label='Default')
axes[1].plot(data, 'g-', drawstyle='steps-post', label='steps-post')

# 設定圖片邊界
plt.ylim([-9, 3])
plt.xlim([-0.5, 10])
```

# 設定主題、軸標籤、刻度和刻度標籤

```python
fig, axes = plt.subplots(3, 1)

data = np.random.randn(1000).cumsum()

### Pic1
axes[0].plot(data, label='Default')

### Pic2
axes[1].plot(data, label='Default')
# x 刻度標示
axes[1].set_xticks([0, 250, 500, 750, 1000])

### Pic3
axes[2].plot(data, label='Default')
axes[2].set_xticks([0, 250, 500, 750, 1000])
# x 刻度標示
labels = axes[2].set_xticklabels(['one', 'two', 'three', 'four', 'five'], rotation=30, fontsize='small')
# 設定主題
axes[2].set_title('This is the title')
# x 刻度標籤
axes[2].set_xlabel('Times')
# 圖例
axes[2].legend(loc='best')

# 子畫面間的距離
plt.subplots_adjust(wspace=0, hspace=1)
```

# 注釋

```python
a = np.arange(0.0,5.0,0.02)
plt.plot(a,np.cos(2*np.pi*a),'r--')
plt.xlabel('Amplitude', fontsize=20, color = 'green')
plt.ylabel('Time', fontsize=20)

# 注釋
plt.annotate(r'$\mu=100$',xy=(2,1),xytext=(3,1.5),
    arrowprops=dict(facecolor='black',shrink=0.1,width=2))

# 網格
plt.grid()
```

# 圖形、存檔

```python
fig = plt.figure()
ax = fig.add_subplot(1, 1, 1)

rect = plt.Rectangle((0.2, 0.75), 0.4, 0.15, color='r', alpha=0.3)
circ = plt.Circle((0.7, 0.2), 0.15, color='g', alpha=0.3)
pgon = plt.Polygon([[0.15, 0.15], [0.35, 0.4], [0.26, 0.6]], color='b', alpha=0.5)

# 加入圖形
ax.add_patch(rect)
ax.add_patch(circ)
ax.add_patch(pgon)

# fname: svg, png, pdf, ps, eps...
# dpi: 預設 100
# bbox_inches: 圖的哪個部份要存檔，'tight': 把圖周圍的空白拿掉
# facecolor: 子圖之外的圖背景色，預設 'w' (white)
# edgecolor
# format: svg, png, pdf, ps, eps...
plt.savefig('pic.png', dpi=400, bbox_inches='tight')

# 存到物件
# from io import BytesIO
# buffer = BytesIO()
# plt.savefig(buffer)
# plot_data = buffer.getvalue()
```

# 其他類型的圖

```python
# 普通圖
from pylab import *

n = 256
X = np.linspace(-np.pi,np.pi,n,endpoint=True)
Y = np.sin(2*X)

plot (X, Y+1, color='blue', alpha=1.00)
plot (X, Y-1, color='blue', alpha=1.00)
```

```python
# 散點圖
from pylab import *

n = 1024
X = np.random.normal(0,1,n)
Y = np.random.normal(0,1,n)

scatter(X,Y)
show()
```

```python
# 條形圖
from pylab import *

n = 12
X = np.arange(n)
Y1 = (1-X/float(n)) * np.random.uniform(0.5,1.0,n)
Y2 = (1-X/float(n)) * np.random.uniform(0.5,1.0,n)

bar(X, +Y1, facecolor='#9999ff', edgecolor='white')
bar(X, -Y2, facecolor='#ff9999', edgecolor='white')

for x,y in zip(X,Y1):
    text(x+0.4, y+0.05, '%.2f' % y, ha='center', va= 'bottom')

ylim(-1.25,+1.25)
```

```python
# 等高線圖
from pylab import *

def f(x,y): return (1-x/2+x**5+y**3)*np.exp(-x**2-y**2)

n = 256
x = np.linspace(-3,3,n)
y = np.linspace(-3,3,n)
X,Y = np.meshgrid(x,y)

contourf(X, Y, f(X,Y), 8, alpha=.75, cmap='jet')
C = contour(X, Y, f(X,Y), 8, colors='black', linewidth=.5)
```

```python
# 灰度圖（Imshow
from pylab import *

def f(x,y): return (1-x/2+x**5+y**3)*np.exp(-x**2-y**2)

n = 10
x = np.linspace(-3,3,4*n)
y = np.linspace(-3,3,3*n)
X,Y = np.meshgrid(x,y)
imshow(f(X,Y))
```

```python
# 餅狀圖
from pylab import *

n = 20
Z = np.random.uniform(0,1,n)
piePics = pie(Z)
```

```python
# 量場圖（Quiver Plots）
from pylab import *

n = 8
X,Y = np.mgrid[0:n,0:n]
quiver(X,Y)
```

```python
# 3D 圖
from pylab import *
from mpl_toolkits.mplot3d import Axes3D

fig = figure()
ax = Axes3D(fig)
X = np.arange(-4, 4, 0.25)
Y = np.arange(-4, 4, 0.25)
X, Y = np.meshgrid(X, Y)
R = np.sqrt(X**2 + Y**2)
Z = np.sin(R)

ax.plot_surface(X, Y, Z, rstride=1, cstride=1, cmap='hot')
```
